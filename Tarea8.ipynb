```python
# Definimos un diccionario 'format_dict' para especificar formatos de visualización personalizados para las columnas.
# 'data science' se muestra con un formato monetario ($0.00)
# 'Mes' se muestra con un formato de fecha (mes y año)
# 'machine learning' se muestra como porcentaje con dos decimales (0.00%)
format_dict = {'data science':'${0:,.2f}', 'Mes':'{:%m-%Y}', 'machine learning':'{:.2%}'}

# Aseguramos que la columna 'Mes' tenga un formato de fecha y hora para su visualización adecuada
df['Mes'] = pd.to_datetime(df['Mes'])

# Aplicamos los estilos de formato definidos en 'format_dict' a las primeras 5 filas del DataFrame.
# Esto formateará la visualización de esas filas según las especificaciones de formato personalizado.
df.head().style.format(format_dict)
```

```python
# Definimos un diccionario de formato simplificado 'format_dict' que se aplica solo a la columna 'Mes'.
# El formato especificado '{:%m-%Y}' muestra la columna 'Mes' en formato de mes y año (por ejemplo, '01-2023').
format_dict = {'Mes':'{:%m-%Y}'}

# Tomamos las primeras 5 filas del DataFrame 'df', aplicamos el formato definido en 'format_dict'.
# Luego, destacamos el valor máximo en color verde oscuro y el valor mínimo en color rojo.
# Esto ayuda a resaltar visualmente los valores extremos en la tabla.
df.head().style.format(format_dict).highlight_max(color='darkgreen').highlight_min(color='#ff0000')
```

```python
# Tomamos las primeras 10 filas del DataFrame 'df'.
# Luego, aplicamos un formato personalizado definido en 'format_dict' a esas filas.
# Esto formateará la visualización de datos según las especificaciones de formato personalizado.

# Además, aplicamos un degradado de fondo (background gradient) en las columnas 'data science' y 'machine learning'.
# El degradado se genera utilizando el mapa de colores 'BuGn', que va desde azul claro hasta verde oscuro.
# Esto permite visualizar la variación de valores en estas dos columnas de manera más visual.
df.head(10).style.format(format_dict).background_gradient(subset=['data science', 'machine learning'], cmap='BuGn')
```

```python
# Tomamos las primeras 5 filas del DataFrame 'df'.
# Luego, aplicamos un formato personalizado definido en 'format_dict' a esas filas.
# Esto formateará la visualización de datos según las especificaciones de formato personalizado.

# Además, aplicamos un estilo de barra (bar) en las columnas 'data science' y 'deep learning'.
# Las barras se colorearán de rojo y su longitud representará los valores en estas columnas.
df.head().style.format(format_dict).bar(color='red', subset=['data science', 'deep learning'])
```

```python
#Matplot
# Importamos la biblioteca Matplotlib y la abreviamos como 'plt' para su uso.
# Luego, creamos un gráfico de línea utilizando los datos de la columna 'Mes' en el eje x
# y los datos de la columna 'data science' en el eje y. Le asignamos una etiqueta 'data science' para la leyenda.
import matplotlib.pyplot as plt
plt.plot(df['Mes'], df['data science'], label='data science') #The parameter label is to indicate the legend. This doesn't mean that it will be shown, we'll have to use another command that I'll explain later.
```

```python
# Creamos tres gráficos de línea en el mismo gráfico.
# Cada llamada a plt.plot() representa una línea en el gráfico.

# El primer plt.plot() traza la columna 'data science' en función de la columna 'Mes'
# y le asigna una etiqueta 'data science' para la leyenda.
plt.plot(df['Mes'], df['data science'], label='data science')
# El segundo plt.plot() traza la columna 'machine learning' en función de la columna 'Mes'
# y le asigna una etiqueta 'machine learning' para la leyenda.
plt.plot(df['Mes'], df['machine learning'], label='machine learning')
# El tercer plt.plot() traza la columna 'deep learning' en función de la columna 'Mes'
# y le asigna una etiqueta 'deep learning' para la leyenda.
plt.plot(df['Mes'], df['deep learning'], label='deep learning')
```

```python
# Creamos tres gráficos de línea en el mismo gráfico.
# Cada llamada a plt.plot() representa una línea en el gráfico.

# El primer plt.plot() traza la columna 'data science' en función de la columna 'Mes'
# y le asigna una etiqueta 'data science' para la leyenda.
plt.plot(df['Mes'], df['data science'], label='data science')

# El segundo plt.plot() traza la columna 'machine learning' en función de la columna 'Mes'
# y le asigna una etiqueta 'machine learning' para la leyenda.
plt.plot(df['Mes'], df['machine learning'], label='machine learning')

# El tercer plt.plot() traza la columna 'deep learning' en función de la columna 'Mes'
# y le asigna una etiqueta 'deep learning' para la leyenda.
plt.plot(df['Mes'], df['deep learning'], label='deep learning')

# A continuación, agregamos etiquetas a los ejes x e y, así como un título al gráfico.
# 'xlabel' establece la etiqueta en el eje x como 'Date' (Fecha).
# 'ylabel' establece la etiqueta en el eje y como 'Popularity' (Popularidad).
# 'title' establece el título del gráfico como 'Popularity of AI terms by date' (Popularidad de términos de IA por fecha).
plt.xlabel('Date')
plt.ylabel('Popularity')
plt.title('Popularity of AI terms by date')

# Luego, habilitamos la cuadrícula en el gráfico con 'grid(True)' para una mejor visualización.
plt.grid(True)

# Finalmente, agregamos una leyenda al gráfico con 'legend()' para identificar las líneas trazadas.
plt.legend()
```

```python
# Creamos una figura (fig) que contiene una matriz de sub-gráficos 2x2 utilizando 'plt.subplots(2, 2)'.
# Esto significa que tendremos 4 sub-gráficos dispuestos en una cuadrícula de 2 filas y 2 columnas.
fig, axes = plt.subplots(2,2)
# En el primer sub-gráfico (axes[0, 0]), creamos un histograma de la columna 'data science'.
axes[0, 0].hist(df['data science'])
# En el segundo sub-gráfico (axes[0, 1]), creamos un gráfico de dispersión ('scatter plot')
# que muestra la relación entre la columna 'Mes' en el eje x y la columna 'data science' en el eje y.
axes[0, 1].scatter(df['Mes'], df['data science'])
# En el tercer sub-gráfico (axes[1, 0]), trazamos una gráfica de línea que representa la columna 'machine learning'
# en función de la columna 'Mes'.
axes[1, 0].plot(df['Mes'], df['machine learning'])
# En el cuarto sub-gráfico (axes[1, 1]), trazamos una gráfica de línea que representa la columna 'deep learning'
# en función de la columna 'Mes'.
axes[1, 1].plot(df['Mes'], df['deep learning'])
```

```python
# En el primer gráfico, trazamos la columna 'data science' en función de la columna 'Mes'
# utilizando un estilo de línea roja continua ('r-').
plt.plot(df['Mes'], df['data science'], 'r-')
# En el segundo gráfico, trazamos la columna 'data science' multiplicada por 2 en función de 'Mes'
# utilizando puntos azules cuadrados ('bs').
plt.plot(df['Mes'], df['data science']*2, 'bs')
# En el tercer gráfico, trazamos la columna 'data science' multiplicada por 3 en función de 'Mes'
# utilizando triángulos verdes ('g^').
plt.plot(df['Mes'], df['data science']*3, 'g^')
```

```python
# Creamos un gráfico de barras ('bar plot') utilizando la columna 'Mes' en el eje x y la columna 'machine learning' en el eje y.
# La opción 'width=20' establece el ancho de las barras en el gráfico.
plt.bar(df['Mes'], df['machine learning'], width=20)
```

```python
#Histograma
# Creamos un histograma utilizando los valores de la columna 'deep learning'.
# La opción 'bins=15' especifica que deseamos dividir los datos en 15 intervalos para construir el histograma.
plt.hist(df['deep learning'], bins=15)
```

```python
#Correccion de codigo

# Graficar las líneas de cada columna
plt.plot(df['Mes'], df['data science'], label='Data Science')
plt.plot(df['Mes'], df['machine learning'], label='Machine Learning')
plt.plot(df['Mes'], df['deep learning'], label='Deep Learning')

plt.xlabel('Fecha')
plt.ylabel('Popularidad')
plt.title('Popularidad de términos de IA por fecha')
plt.grid(True)

# Agregar texto y anotación al gráfico
plt.text(df['Mes'].min(), 80, r'$\lambda=1, r^2=0.8$')
plt.annotate('¿Notas algo importante?', xy=(pd.to_datetime('2014-01-01'), 30), xytext=(pd.to_datetime('2006-01-01'), 50), arrowprops={'facecolor':'red', 'shrink':0.05})

# Formatear las fechas en el eje x para que se vean mejor
plt.xticks(rotation=45)

# Mostrar el gráfico
plt.legend()
plt.tight_layout()  # Ajusta el diseño para evitar cortes en etiquetas
plt.show()
```

```python
import seaborn as sns
# Corrige los nombres de las columnas según corresponda
sns.relplot(x='Mes', y='deep learning', hue='data science', size='machine learning', data=df)
```

```python
# Seleccionar solo las columnas numéricas para la matriz de correlación
num_cols = ['data science', 'machine learning', 'deep learning']
# Convertir las columnas numéricas a valores numéricos si es necesario
for col in num_cols:
    df[col] = pd.to_numeric(df[col], errors='coerce')
# Obtener la matriz de correlación solo para las columnas numéricas
correlation_matrix = df[num_cols].corr()
# Generar el heatmap
sns.heatmap(correlation_matrix, annot=True, fmt='.2f')
```

```python
# Creamos un gráfico de pares ('pair plot') utilizando el DataFrame 'df'.
# Este gráfico muestra relaciones entre pares de variables en 'df', incluyendo gráficos de dispersión y histogramas.
sns.pairplot(df)
```

```python
#Graficos segmentados segun variable categorica
# Creamos un gráfico de pares ('pair plot') utilizando el DataFrame 'df' y utilizando 'categorical' como variable categórica para colorear puntos.
# Este gráfico muestra relaciones entre pares de variables en 'df' y utiliza colores diferentes para visualizar categorías de la variable 'categorical'.
sns.pairplot(df, hue='categorical')
```

```python
# Creamos un gráfico de dispersión con histogramas marginales ('joint plot') utilizando las columnas 'data science' y 'machine learning' del DataFrame 'df'.
# Este gráfico muestra la relación entre estas dos variables, incluyendo histogramas en los márgenes para visualizar sus distribuciones individuales.
sns.jointplot(x='data science', y='machine learning', data=df)
```

```python
# Creamos un gráfico de violín ('violin plot') utilizando la columna 'categorical' en el eje x y la columna 'data science' en el eje y del DataFrame 'df'.
# Este gráfico muestra la distribución de los valores de 'data science' dentro de cada categoría de la variable 'categorical'.
sns.catplot(x='categorical', y='data science', kind='violin', data=df)
```

```python
# Creamos una figura (fig) que contiene una fila con dos sub-gráficos (axes) compartiendo el eje y ('sharey=True') y con un tamaño de figura de 8x4 ('figsize=(8, 4)').
fig, axes = plt.subplots(1, 2, sharey=True, figsize=(8, 4))
# En el primer sub-gráfico (axes[0]), creamos un gráfico de dispersión ('scatter plot') que muestra la relación entre 'Mes' en el eje x y 'deep learning' en el eje y.
# Utilizamos el atributo 'hue' para colorear los puntos según la variable 'categorical'.
# Establecemos el título del sub-gráfico como 'Deep Learning'.
sns.scatterplot(x="Mes", y="deep learning", hue="categorical", data=df, ax=axes[0])
axes[0].set_title('Deep Learning')
# En el segundo sub-gráfico (axes[1]), creamos otro gráfico de dispersión que muestra la relación entre 'Mes' en el eje x y 'machine learning' en el eje y.
# Al igual que en el primer sub-gráfico, utilizamos el atributo 'hue' para colorear los puntos según la variable 'categorical'.
# Establecemos el título del sub-gráfico como 'Machine Learning'.
sns.scatterplot(x="Mes", y="machine learning", hue="categorical",
data=df, ax=axes[1])
axes[1].set_title('Machine Learning')
```

```python
# Creamos un mapa ('m2') utilizando Folium, con la ubicación central en [41.38, 2.17].
# Especificamos el tipo de mapa ('tiles') como 'openstreetmap' y el nivel de zoom inicial ('zoom_start') como 16.
m2 = folium.Map(location=[41.38, 2.17], tiles='openstreetmap',zoom_start=16)
# Agregamos dos marcadores al mapa utilizando 'folium.Marker()'.
# El primer marcador tiene una ubicación específica ([41.38, 2.176]), un mensaje emergente con formato HTML y un tooltip.
folium.Marker([41.38, 2.176], popup='<i>You can use whatever HTML code you want</i>', tooltip='click here').add_to(m2)
# El segundo marcador tiene una ubicación diferente ([41.38, 2.174]), otro mensaje emergente con formato HTML y un tooltip diferente.
folium.Marker([41.38, 2.174], popup='<b>You can use whatever HTML code you want</b>', tooltip='dont click here').add_to(m2)
# Guardamos el mapa como un archivo HTML llamado 'map2.html' utilizando el método 'save()'.
m2.save('map2.html')
# Finalmente, mostramos el mapa 'm2' en la salida para visualización interactiva.
m2
```
