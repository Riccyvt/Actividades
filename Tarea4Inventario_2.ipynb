```python
# Crear un diccionario vacío para el inventario
inventario = dict()

# Definir una función para agregar un producto al inventario
def add_prodNvo(nom, cant):
    if nom in inventario.keys():
        inventario[nom] += cant
    else:
        inventario[nom] = cant

# Definir una función para buscar un producto en el inventario
def busq_prodNvo(prod):
    if prod in inventario.keys():
        print(f"El producto {prod} está en el inventario con una cantidad de {inventario[prod]}.")
    else:
        print(f"El producto {prod} no existe en el inventario.")

# Definir una función para actualizar la cantidad de un producto en el inventario
def act_prodNvo(prod, cantidad):
    if prod in inventario.keys():
        inventario[prod] = cantidad
        print(f"Se ha actualizado la cantidad de {prod} a {cantidad}.")
    else:
        print(f"El producto {prod} no existe en el inventario.")

# Definir una función para eliminar un producto del inventario
def borra_prodNvo(nom):
    if nom in inventario:
        del inventario[nom]
        print(f"El producto {nom} ha sido eliminado del inventario.")
    else:
        print("El producto a eliminar no existe en el inventario.")

# Iniciar un bucle infinito para el menú de gestión del inventario
while True:
    print('Menú para gestionar inventario: ')
    print('1 - Agregar producto')
    print('2 - Buscar producto')
    print('3 - Actualizar cantidad')
    print('4 - Eliminar producto')
    print('5 - Salir')

    # Obtener la opción del usuario
    opcion = int(input('Elige la opción a realizar para el inventario: '))

    if opcion == 1:
        # Agregar un producto al inventario
        nom = input('Introduce el nombre del producto: ')
        cant = int(input('Indica la cantidad de producto: '))
        add_prodNvo(nom, cant)
        print(inventario)
    elif opcion == 2:
        # Buscar un producto en el inventario
        prod = input('Introduce el producto a buscar: ')
        busq_prodNvo(prod)
    elif opcion == 3:
        # Actualizar la cantidad de un producto en el inventario
        prod = input('Introduce el producto a modificar: ')
        cant = int(input('Introduce la nueva cantidad: '))
        act_prodNvo(prod, cant)
    elif opcion == 4:
        # Eliminar un producto del inventario
        prod = input('Introduce el producto a borrar: ')
        borra_prodNvo(prod)
    elif opcion == 5:
        # Salir del programa
        print('Saliendo del programa.')
        break
    else:
        print('Opción no válida. Por favor, elige una opción del menú.')

```
